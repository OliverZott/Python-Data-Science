#  Seaborn - Categorical

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import matplotlib
%matplotlib inline

# gloal figsize
matplotlib.rcParams['figure.figsize'] = (12.0, 8.0)

tips = sns.load_dataset("tips")
tips.head()
```

# Boxplot

```python
plt.tight_layout();

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(7,3), dpi=100)

sns.barplot(ax=axes[0], x='sex', y='total_bill', data=tips, )
sns.barplot(ax=axes[1], x='sex', y='total_bill', data=tips, estimator=np.std)
```

```python
sns.countplot(x='sex', data=tips)
```

```python
plt.figure(figsize=(10,6))
plt.grid(b=True, which='both')

sns.boxplot(x='day', y='total_bill', data=tips, palette='rainbow')
```

```python
plt.figure(figsize=(12,7))
plt.grid(b=True)

sns.boxplot(x="day", y="total_bill", hue="smoker", data=tips, palette="coolwarm")
```

# Violin-Plot / Stripl-Plot / Swarm-Plot

Difference to boxplot: shows not actual datapoints as in boxplot, but kde- distributions

```python
sns.violinplot(x="day", y="total_bill", data=tips, hue="sex", palette='rainbow', split=True)
```

```python
sns.stripplot(x="day", y="total_bill", data=tips, jitter=True, hue="sex", palette='Set1', dodge=True)
```

```python
sns.swarmplot(x="day", y="total_bill", data=tips, hue="sex", palette='Set1',)
```

# Combination-Plots

```python
plt.figure(figsize=(15,10))
sns.violinplot(x="day", y="tip", data=tips, palette='rainbow')
sns.swarmplot(x="day", y="tip", data=tips, color='black')
```

# Factor-Plot

Most generic plot with type as parameter (e.g. violin, box,...)

```python
sns.factorplot(x='sex', y='total_bill', data=tips, kind='strip')

# Depricated: new is "catplot"
```
