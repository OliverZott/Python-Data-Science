```python
import numpy as np
import matplotlib.pyplot as plt

# https://ipython.readthedocs.io/en/stable/interactive/plotting.html
%matplotlib inline
```

```python
x = np.linspace(0, 5, 41)
y = x**2

plt.plot(x,y, 'r')
plt.xlabel('x-Label')
plt.ylabel('y-Label')
plt.title('Title')
# plt.show()  # not needed in jupyter (cause "inline")
```

### Subplots

```python
# (rows, columns, position)

plt.subplot(1,2,1)
plt.plot(x, y, 'r--')
plt.subplot(1,2,2)
plt.plot(y,x, 'b-*')
```

### Object Oriented Plot

```python
af = plt.figure()

axes = af.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x, y, 'g--')
axes.set_xlabel('x-label')
axes.set_ylabel('y-label')
axes.set_title('Title')
```

**Advantage of OO plot**

```python
af = plt.figure()

axes1 = af.add_axes([0.1, 0.1, 0.8, 0.9])
axes2 = af.add_axes([0.8, 0.1, 0.2, 0.2])

axes1.plot(x,y, 'b')
axes1.set_xlabel('x-label')
axes1.set_ylabel('y-label')
axes1.set_title('big')

axes2.plot(y,x, 'r')
axes2.set_xlabel('x-label')
axes2.set_ylabel('y-label')
axes2.set_title('small')
```

## Subplots object oriented - Diagrams

- axes are an array
- `tight_layout`
- `figuresize`

```python

diag, axes = plt.subplots(nrows = 1, ncols = 2)

# "axes" is an array! Therefore one can iterate over it
print(axes)


for ax in axes:
    ax.plot(x, y)
    ax.set_xlabel('x-axes')
    ax.set_ylabel('y-axes') 
    plt.tight_layout()                # for broader layout
    
axes[0].plot(x,y, 'r*')
axes[0].set_title('title1')
axes[1].plot(y, x, 'g')
```

```python
# figure size


# https://stackoverflow.com/questions/47633546/relationship-between-dpi-and-figure-size/47639545

diag = plt.figure(figsize=(8,4), dpi=100)

ax = diag.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
diag, axes = plt.subplots(nrows = 1, ncols = 2, figsize=(12,4), dpi=100)

axes[0].plot(y,x)
axes[0].set_title('x^-2')
axes[1].plot(x,y, 'r--')
axes[1].set_title('x^2')


for ax in axes:
    ax.set_xlabel('x-axes')
    ax.set_ylabel('y-axes')
    

# Save diagrams:
diag.savefig("testFig1.png", dpi=200)
diag.savefig("testFig2.eps", dpi=100)
```

## Legends

```python

diag = plt.figure()

ax_leg = diag.add_axes([0,0,1,1])
ax_leg.plot(x, x**2, label="x**2")
ax_leg.plot(x, x**3, label="x**3")
ax_leg.legend(loc=9)
```

## Colors & Limits

```python
diag, ax = plt.subplots(dpi=150)

ax.plot(x, x**2, 'b--', linewidth=0.7, marker='2', markersize=5)
ax.plot(x, x**(-1), color="green", alpha=0.3, lw=0.9, linestyle='-')  # RGB HexCode
ax.set_xlim(-1,6)
ax.set_ylim([0,12])
```
