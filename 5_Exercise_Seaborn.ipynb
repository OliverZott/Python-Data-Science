# Seaborn Übung - Aufgabe

## Die Daten
Wir werde dazu ein berühmten Datensatz zur Titanic benutzen. Später im Machine Learning Teil des Kurses kommen wir auf diesen Datensatz zurück, um Überlebenswahrscheinlichkeiten zu berechnen.

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
titanic = sns.load_dataset('titanic')
titanic.info()
```

```python
titanic.head()
```

## Übung

```python
# Joint PLot
sns.jointplot(x="fare", y="age", data=titanic)
```

```python
sns.distplot(titanic['fare'], bins=30, kde=True)
```

```python
sns.boxplot(x='class', y='age', data=titanic, palette='rainbow')
```

```python
sns.swarmplot(x='class', y='age', size=3, data=titanic, dodge=True, palette='Set2', edgecolor="black",alpha=.5, linewidth=1.0)
```

```python
sns.countplot(data=titanic, x='sex')
```

```python
sns.heatmap(titanic.corr(), cmap='coolwarm')
```

```python
plt.style.use('seaborn')
g = sns.FacetGrid(data=titanic, col='sex')
g.map(sns.displot,'age')
g.map(sns.kdeplot,'age')
```

```python
g = sns.FacetGrid(data=titanic, col='sex')
g.map(sns.distplot,'age')
```
